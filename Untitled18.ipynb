```
df = df.set_index('date')
df['pm25'] = df['pm25'].replace('-',np.nan)
df['pm25']=df['pm25'].astype(str).astype(float)
```

```
df = df.assign(Time=df.pm25.fillna(df.pm25.rolling(50,min_periods=1,).mean()))
```

```
df=df.reindex(index=df.index[::-1])
```

```
df
```

```
df.isnull().sum()
```

```
train=df[0:1900] 
test=df[1900:]
```

```
train
```

```
test
```

```
df.Timestamp = pd.to_datetime(df.index,format='%d-%m-%Y %H:%M') 
df.index = df.Timestamp 
df = df.resample('D').mean()
train.Timestamp = pd.to_datetime(train.index,format='%d-%m-%Y %H:%M') 
train.index = train.Timestamp 
train = train.resample('D').mean() 
test.Timestamp = pd.to_datetime(test.index,format='%d-%m-%Y %H:%M') 
test.index = test.Timestamp 
test = test.resample('D').mean()
```

```
train.isna().sum()
```

```
#Plotting data
train.Time.plot(figsize=(15,8), title= 'Pollution', fontsize=14)
test.Time.plot(figsize=(15,8), title= 'Pollution', fontsize=14)
plt.show()
```

```
dd= np.asarray(train.Time)
y_hat = test.copy()
y_hat['naive'] = dd[len(dd)-1]
plt.figure(figsize=(30,8))
plt.plot(train.index, train['Time'], label='Train')
plt.plot(test.index,test['Time'], label='Test')
plt.plot(y_hat.index,y_hat['naive'], label='Naive Forecast')
plt.legend(loc='best')
plt.title("Naive Forecast")
plt.show()
```

```
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(test.Time, y_hat.naive))
print('RMSE with naive approach : ', rms)
```

```
y_hat_avg = test.copy()
y_hat_avg['avg_forecast'] = train['Time'].mean()
plt.figure(figsize=(12,8))
plt.plot(train['Time'], label='Train')
plt.plot(test['Time'], label='Test')
plt.plot(y_hat_avg['avg_forecast'], label='Average Forecast')
plt.legend(loc='best')
plt.show()
```

```
rms = sqrt(mean_squared_error(test.Time, y_hat_avg.avg_forecast))
print('RMSE with Simple Average approach : ', rms)

```

```
y_hat_avg = test.copy()
y_hat_avg['moving_avg_forecast'] = train['Time'].rolling(60).mean().iloc[-1]
plt.figure(figsize=(16,8))
plt.plot(train['Time'], label='Train')
plt.plot(test['Time'], label='Test')
plt.plot(y_hat_avg['moving_avg_forecast'], label='Moving Average Forecast')
plt.legend(loc='best')
plt.show()
```

```
rms = sqrt(mean_squared_error(test.Time, y_hat_avg.moving_avg_forecast))
print('RMSE with Moving Average approach : ', rms)
```

```
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
y_hat_avg = test.copy()
fit2 = SimpleExpSmoothing(np.asarray(train['Time'])).fit(smoothing_level=0.6,optimized=False)
y_hat_avg['SES'] = fit2.forecast(len(test))
plt.figure(figsize=(16,8))
plt.plot(train['Time'], label='Train')
plt.plot(test['Time'], label='Test')
plt.plot(y_hat_avg['SES'], label='SES')
plt.legend(loc='best')
plt.show()
```

```
rms = sqrt(mean_squared_error(test.Time, y_hat_avg.SES))
print('RMSE with Simple Exponential Smoothing approach : ', rms)
```

```
import statsmodels.api as sm
sm.tsa.seasonal_decompose(train.Time).plot()
result = sm.tsa.stattools.adfuller(train.Time)
plt.show()
```

```
y_hat_avg = test.copy()

fit1 = Holt(np.asarray(train['Time'])).fit()
y_hat_avg['Holt_linear'] = fit1.forecast(len(test))

plt.figure(figsize=(16,8))
plt.plot(train['Time'], label='Train')
plt.plot(test['Time'], label='Test')
plt.plot(y_hat_avg['Holt_linear'], label='Holt_linear')
plt.legend(loc='best')
plt.show()
```

```
rms = sqrt(mean_squared_error(test.Time, y_hat_avg.Holt_linear))
print('RMSE with Holt Linear approach : ', rms)
```

```
y_hat_avg = test.copy()
fit1 = ExponentialSmoothing(np.asarray(train['Time']) ,seasonal_periods=7 ).fit()
y_hat_avg['Holt_Winter'] = fit1.forecast(len(test))
plt.figure(figsize=(16,8))
plt.plot( train['Time'], label='Train')
plt.plot(test['Time'], label='Test')
plt.plot(y_hat_avg['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
plt.show()
```

```
rms = sqrt(mean_squared_error(test.Time, y_hat_avg.Holt_Winter))
print('RMSE with Holt Winter approach : ', rms)
```
