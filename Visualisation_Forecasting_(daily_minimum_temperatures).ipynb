#### Box and Whisker Plots by Interval

```python
# create a boxplot of yearly data
from pandas import read_csv
from pandas import DataFrame
from pandas import Grouper
from matplotlib import pyplot
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,squeeze=True,parse_dates=True)
groups = series.groupby(Grouper(freq='A'))
years = DataFrame()
for name, group in groups:
    years[name.year] = group.values
years.boxplot()
pyplot.show()
```

1981, 1982, 1984, 1986, 1987, 1988 has outliers

#### Lag plot

```python
# create a scatter plot
from pandas import read_csv
from matplotlib import pyplot
from pandas.plotting import lag_plot
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True)
lag_plot(series, lag=2)
pyplot.show()
```

```python
# create an autocorrelation plot
from pandas import read_csv
from matplotlib import pyplot
from statsmodels.graphics.tsaplots import plot_acf
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True)
plot_acf(series,lags=10)
pyplot.show()
```

Every Lag is showing significance
