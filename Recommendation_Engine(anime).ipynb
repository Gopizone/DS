```
import pandas as pd
import numpy as np
```

```
anime_df = pd.read_csv('anime.csv')
```

```
anime_df
```

```
#number of unique users in the dataset
len(anime_df.anime_id.unique())
```

```
len(anime_df.name.unique())
```

```
len(anime_df.genre.unique())
```

```
user_anime_df = anime_df.pivot(index='anime_id',
                                 columns='genre',
                                 values='rating').reset_index(drop=True)
```

```
user_anime_df
```

```
user_anime_df.index = anime_df.anime_id.unique()
```

```
user_anime_df
```

```
#Impute those NaNs with 0 values
user_anime_df.fillna(0, inplace=True)
```

```
user_anime_df
```

```
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```
user_sim = 1 - pairwise_distances( user_anime_df.values,metric='cosine')
```

```
user_sim
```

```
user_sim.shape
```

```
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```
#Set the index and column names to user ids 
user_sim_df.index = anime_df.anime_id.unique()
user_sim_df.columns = anime_df.anime_id.unique()
```

```
user_sim_df.iloc[0:5, 0:5]
```

```
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```
